```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# Q(1) -----------------------------------------------------------------------------------

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
driver.get(" https://www.naukri.com/")
driver.find_element_by_class_name('suggestor-input').send_keys('Data Analyst')
driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input').send_keys('Banglore')
driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]').click()
jobt=driver.find_elements(By.XPATH,"//a[contains(@class,'title fw500 ellipsis')]")
city=driver.find_elements(By.XPATH,"//li[contains(@class,'fleft grey-text br2 placeHolderLi location')]")
comname=driver.find_elements(By.XPATH,"//a[contains(@class,'subTitle ellipsis fleft')]")
ex=driver.find_elements(By.XPATH,"//li[contains(@class,'fleft grey-text br2 placeHolderLi experience')]")
job_title=[]
location=[]
comp_name=[]
experience=[]
for title in jobt:
    print(title.text)
    job_title.append(title.text)
    
print("@"*50)

for loc in city:
    print(loc.text)
    location.append(loc.text)
print("#"*50)

for comp in comname:
    print(comp.text)
    comp_name.append(comp.text)
print("%"*50)

for exp in ex:
    print(exp.text)
    experience.append(exp.text)

len(job_title), len(location), len(comp_name), len(experience)

df=pd.DataFrame()
df['Job Title']=job_title
df['Location']=location
df['Company Name']=comp_name
df['Experience']=experience
df[0:10]
```

```python
driver=webdriver.Chrome(r'/Users/nimeshroy/Downloads/chromedriver')
driver.get('https://www.myntra.com/shoes')
driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div').click()
#driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div').click()
```

```python
name=[]
for i in range(2):
    print("scraping page no. - ", i+1)
    na=driver.find_elements(By.XPATH,'//h3[contains(@class,"product-brand")]')
    for v in na:
        a=v.text.replace('','')
        name.append(a.replace('',''))
    click_next=driver.find_element_by_xpath('//li[@class="pagination-next"]')
    click_next.click()
    sleep(2)
len(name)
```

```python
shortd=[]
for i in range(2):
    print("scraping page no. - ", i+1)
    sd=driver.find_elements(By.XPATH,'//h4[contains(@class,"product-product")]')
    for w in sd:
        b=w.text.replace('','')
        shortd.append(b.replace('',''))
    click_next=driver.find_element_by_xpath('//li[@class="pagination-next"]')
    click_next.click()
    sleep(2)
len(shortd)
```

```python
price=[]
for i in range(2):
    print("scraping page no. - ", i+1)
    pr=driver.find_elements(By.XPATH,'//div[contains(@class,"product-price")]')
    for u in pr:
        c=u.text.replace('','')
        price.append(c.replace('',''))
    click_next=driver.find_element_by_xpath('//li[@class="pagination-next"]')
    click_next.click()
    sleep(2)
len(price)
```

```python
df=pd.DataFrame()
df["Brand"]=name
df["Short Description"]=shortd
df["Price"]=price
df
```
